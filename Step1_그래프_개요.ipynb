### 1.4 Graph Basic Concepts

 - 노드의 차수(degree of node) : 노드에 입사하는 엣지의 수
 - 노드가 엣지의 끝점 중 하나이면 엣지가 노드에 인시던트되어 있다고 함
 - 노드의 정도(v) 는 deg(v)로 표시함

 - 방향이 없는 그래프에서 정점의 차수는 정점에 연결된 가장자리 수. 노드가 루프 또는 자기 루프라고 불리는 노드 자체에 연결되어 있으면 차수에 2가 더해짐
 - 방향 그래프에서 차수는 내차수(indegree)와 외차수(outdegree)의 두 가지 유형으로 나뉘며 노드의 내차수 deg-(v)로 표시는 해당 노드를 가리키는 엣지의 수를 나타내고, 외차수 deg+(v)로 표시는 해당 노드에서 시작하는 에지의 수를 나타냄. 이 경우 자체 루프는 차수와 차외 차수에각각 하나를 더함
---
==> 풀어서 얘기하면

 - Degree of Node 란? : 노드에 연결된 엣지의 수를 의미
 - 즉, 하나의 노드에 엣지가 몇 개나 붙어 있는지! 
 - 인시던트(incident)란? : "엣지가 어떤 노드에 연결되어 있다"는 것을 "엣지가 그 노드에 인시던트되어 있다"고 표현
 - deg(v) 란? : 노드 v의 차수를 수학적으로 나타낸 경우. 노드 v의 차수가 2라면 deg(v) = 2
 - 루프(loop) 란? : 노드가 자기 자신과 연결된 경우.
 A --- A 의 경우 A에 루프 하나가 생기는데 이 루프는 시작점도 A, 끝점도 A 이므로 엣지가 두 번 연결된 걸로 봐서 차수는 2 증가
 - A -> B 의 Directed Graph 에서 A는 outdegree(차외 차수) 1(엣지가 나감), B는 indegree(차내 차수) 1(엣지가 들어옴)
 - dev^-(A) = 0, dev^+(A) = 1 
 - dev^-(B) = 1, dev^+(B) = 0

```python
# 노드의 차수 계산 방법
G = nx.Graph()
G.add_edges_from([("A", "B"), ("A", "C"), ("B", "D"), ("B", "E"), ("C", "F"), ("C", "G")])
print(f"deg(A) = {G.degree['A']}")

DG = nx.DiGraph()
DG.add_edges_from([('A','B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G')])
print(f"deg^-(A) = {DG.in_degree['A']}")
print(f"deg^+(A) = {DG.out_degree['A']}")
```

### Graph measure

 - 그래프에서 Degree와 Path는 노드의 중요도를 결정하는 정량 지표로 사용됨. --> 이 측정값을 중심성(Centrality) 라고 함.

노드의 중요도 측정 방법은 다음과 같음.
 1. 차수 중심성(Degree Centrality)
 2. 근접 중심성(Closeness Centrality)
 3. 매개 중심성(Betweenness Centrality)

##### 1. 차수 중심성(Degree Centrality)
 - 특정 노드가 얼마나 많은 다른 노드와 직접적으로 연결되어 있는지 나타 냄.
 - 이는 노드의 차수로 정의 됨.
 - 높은 차수 중심성을 가진 노드는 다른 노드들과 많이 연결되어 있어 네트워크에서 중요한 역할을 함.
 - ex) 네트워크에서 차수 중심성이 높은 사람은 많은 친구나 팔로워를 가진 사람으로, 정보나 영향력이 빠르게 확산될 가능성이 높음.

 ==> 노드의 직접적연 연결성을 평가

##### 2. 근접 중심성(Closeness Centrality)
 - 특정 노드가 다른 모든 노드에 얼마나 가까운지를 측정하는 척도.
 - 해당 노드와 나머지 모든 노드 간 최단 경로 길이의 평균값으로 계산됨.
 - 높은 근접 중심성을 가진 노드는 네트워크에서 다른 모든 노드에 빠르게 도달할 수 있어, 정보 전달이나 네트워크 내 이동이 효율적인 노드로 평가됨.
 - 즉, 네트워크 중심부에 위치한 노드를 식별하는 데 유용함.

 ==> 노드가 다른 노드와의 거리에 기반하여 평가

##### 3. 매개 중심성(Betweenness Centrality)
 - 특정 노드가 다른 노드 쌍 간의 최단 경로에 얼마나 자주 포함되는지를 측정하는 척도.
 - 높은 매개 중심성을 가진 노드는 네트워크에서 서로 다른 영역을 연결하는 다리 또는 중개자 역할을 함.
 - 정보 흐름을 조정하거나 제어할 수 있는 중요한 위치에 있음.
 - 이러한 노드는 네트워크의 병목 지점으로 작용할 수 있음.

 ==> 네트워크 내 중개자 역할의 중요성을 평가

##### + Eigenvector 중심성(Eigenvector Centrality)
 - 단순히 연결 수만 보는 것이 아니라, 영향력 있는 이웃과 연결된 노드일수록 더 중요하다는 내용을 반영한 그래프 중심성 지표.
 - 즉, 중요한 노드와 연결된 노드는 더 중요하게 평가함.
 - ex) 연결된 노드의 수가 동일하더라도 그 중 더 중요한 노드와 연결된 노드를 더 중요하다고 평가함.
 - 영향력 전파 중심을 강조함

### 그래프 알고리즘
 - 두 노드 사이의 최단 경로를 찾거나 주기를 감지하는 등 그래프와 관련된 문제를 해결하는 데 중요.
 - BFS(Breadth First Search)와 DFS(Depth First Search)

##### BFS vs DFS
 - 탐색 순서 : BFS = 시작 노드에서 가까운 노드부터 탐색, DFS = 한 경로를 따라 가능한 깊이까지 탐색
 - 자료 구조 : BFS = 큐(Queue), DFS = 스택(Stack) 또는 재귀 호출(Recursive Call)
 - 응용 사례 : BFS = 최단 경로 탐색과 계층적 분석에 적합, DFS = 경로 탐색, 순환 여부 확인, 위상 정렬에 적합
 - 시간 복잡도 : BFS와 DFS 모두 O(V + E)의 시간 복잡도를 가짐

```python
# BFS(Breadth-first search)

G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G')])

def bfs(graph, node):
    visited, queue = [node], [node]

    while queue:
        node = queue.pop(0)
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.append(neighbor)
                queue.append(neighbor)
    return visited
```
